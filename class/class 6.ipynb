```python
class ExpenseTracker:
    def _init_(self):
        self.data = []

    def add_transaction(self, expense_type, category, cost, description, date):
        self.data.append({"type": expense_type, "category": category, "cost": cost, "description": description, "date": date})

    def total_expense(self):
        total = 0
        for transaction in self.data:
            if transaction["type"].lower() == "expense":
                total += float(transaction["cost"])
        return total

    def total_income(self):
        total = 0
        for transaction in self.data:
            if transaction["type"].lower() == "income":
                total += float(transaction["cost"])
        return total

def save_to_csv(data, filename):
    with open(filename, 'w') as csvfile:
        csvfile.write("type,category,cost,description,date\n")
        for transaction in data:
            row = ",".join([transaction["type"], transaction["category"], str(transaction["cost"]), transaction["description"], transaction["date"]])
            csvfile.write(row + "\n")

def load_from_csv(filename):
    tracker = ExpenseTracker()
    try:
        with open(filename, 'r') as csvfile:
            next(csvfile)  # Skip the header row
            for line in csvfile:
                parts = line.strip().split(",")
                expense_type = parts[0]
                category = parts[1]
                cost = float(parts[2])
                description = parts[3]
                date = parts[4]
                tracker.add_transaction(expense_type, category, cost, description, date)
    except FileNotFoundError:
        pass  # File doesn't exist yet

    return tracker

def main():
    filename = "expense.csv"
    tracker = load_from_csv(filename)

    while True:
        print("\nMenu:")
        print("1. Add Expense")
        print("2. Add Income")
        print("3. View Total Expense")
        print("4. View Total Income")
        print("5. Save and Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            expense_type = "Expense"
            category = input("Enter the expense category: ")
            cost = float(input("Enter the expense cost: "))
            description = input("Enter a description: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            tracker.add_transaction(expense_type, category, cost, description, date)
        elif choice == '2':
            expense_type = "Income"
            category = input("Enter the income category: ")
            cost = float(input("Enter the income amount: "))
            description = input("Enter a description: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            tracker.add_transaction(expense_type, category, cost, description, date)
        elif choice == '3':
            total_expense = tracker.total_expense()
            print(f"Total Expense: {total_expense}")
        elif choice == '4':
            total_income = tracker.total_income()
            print(f"Total Income: {total_income}")
        elif choice == '5':
            save_to_csv(tracker.data, filename)
            print("Data saved. Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")

if _name_ == "_main_":
    main()
```

```python
import csv


class expenseTracker:
    def _init_(self, csv_filename):
        self.expenseDict = {
            "income": [],
            "expense": [],
        }
        self.csv_filename = csv_filename

        # Create the CSV file and write the header if it doesn't exist
        with open(self.csv_filename, mode="a+", newline="") as file:
            writer = csv.writer(file)
            # Check if the file is empty and write the header if needed
            if file.tell() == 0:
                writer.writerow(["Amount", "Category", "Date", "Details", "Type"])

    def store_transactions(self, type, amt, category, date, details):
        trans = {
            "Amount": amt,
            "Category": category,
            "Date": date,
            "Details": details,
        }
        if type == "income":
            self.expenseDict["income"].append(trans)
            transaction_type = "Income"
        else:
            self.expenseDict["expense"].append(trans)
            transaction_type = "Expense"

        # Add the transaction to the CSV file with the transaction type
        self.add_transaction_to_csv(amt, category, date, details, transaction_type)

    def view_transactions(self):
        print("Your Income:")
        for item in self.expenseDict["income"]:
            print(item)
        print("Your Expenses:")
        for item in self.expenseDict["expense"]:
            print(item)

    def calculate_transactions(self):
        total_income = sum(item["Amount"] for item in self.expenseDict["income"])
        total_expense = sum(item["Amount"] for item in self.expenseDict["expense"])
        print("Total Income\t:\t", total_income)
        print("Total Expenses\t:\t", total_expense)

    def add_transaction_to_csv(self, amt, category, date, details, transaction_type):
        with open(self.csv_filename, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([amt, category, date, details, transaction_type])


def collectInput():
    amt = int(input("Enter the amount: "))
    category = input("Enter Category: ")
    date = input("Enter Date (DD/MM/YYYY): ")
    details = input("Enter description: ")
    return amt, category, date, details


myexpense = expenseTracker("expense_records.csv")

while True:
    print("...MY EXPENSE TRACKER...")
    print("1. Record Income")
    print("2. Record Expense")
    print("3. View Records")
    print("4. View My Spending")
    print("5. Exit")

    choice = int(input("Enter your choice: ").strip())

    if choice == 1:
        print("Enter the details of income")
        amt, category, date, details = collectInput()
        myexpense.store_transactions("income", amt, category, date, details)
    elif choice == 2:
        print("Enter the details of expense")
        amt, category, date, details = collectInput()
        myexpense.store_transactions("expense", amt, category, date, details)
    elif choice == 3:
        myexpense.view_transactions()
    elif choice == 4:
        myexpense.calculate_transactions()
    elif choice == 5:
        exit()
    else:
        print("Invalid choice")


```
