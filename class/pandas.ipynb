```python
#read csv file and print the n rows
#class pandas_Name
     #init()
          #stores the data
    #read_csv()
          #reads a csv file
    #head()
          #prints the top n rows
    #tail()
          #prints the bottom n rows

class pandas_dipanwita:
      def __init__(self):
            self.filename = ""
            self.filecontent=""
            
      def read_csv(self,filename):
            self.filename = filename
            try:
               with open(filename,'r') as file:    #to check error
                  self.filecontent =file.readlines()
            except:
                  print("file not found")
      def head(self,num_of_rows):
            for i in range(0,num_of_rows):
                  print(self.filecontent[i].strip())
            #tnp =self.filecontent[num]
            #for item in tnp:
                  #print(item.strip())
      
      def tail(self,num_of_rows):
            start = len(self.filecontent)-(num_of_rows+1)
            end = len(self.filecontent) -1
            for i in range(start,end):
                  print(self.filecontent[i].strip())
        
        
    
```

```python
pd = pandas_dipanwita()
pd.read_csv("sample.csv")
pd.head(2)
pd.tail(2)
```

```python
class pandas_dipanwita:
      def __init__(self):
            self.filename = ""
            self.filecontent=""
            self.columns =""
      def read_csv(self,filename):
            self.filename = filename
            try:
               with open(filename,'r') as file: #to check error
                  self.columns = file.readline()
                  self.filecontent =file.readlines()
            except:
                  print("file not found")
      def head(self,num_of_rows):
            for i in range(0,num_of_rows):
                  print(self.filecontent[i].strip())
            #tnp =self.filecontent[num]
            #for item in tnp:
                  #print(item.strip())
      
      def tail(self,num_of_rows):
            start = len(self.filecontent)-(num_of_rows+1)
            end = len(self.filecontent) -1
            for i in range(start,end):
                  print(self.filecontent[i].strip())
        
```

```python
pd= pandas_dipanwita()
pd.read_csv("sample.csv")
pd.columns
```
