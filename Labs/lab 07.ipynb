```python
import random

# Predefined menu items and their prices
menu = [["Rice", 30], ["Tea", 10], ["Pasta", 40], ["Salad", 20]]

# Data storage for orders
orders = []

def create_new_order():
    order_id = generate_order_id()
    items = []
    
    print("Menu:")
    for i, (item, price) in enumerate(menu, 1):
        print(f"{i}. {item} - ${price}")
    
    while True:
        item_number = input("Enter item number to add to the order (or 'done' to finish): ")
        if item_number == 'done':
            break
        try:
            item_number = int(item_number)
            if 1 <= item_number <= len(menu):
                quantity = int(input("Enter quantity: "))
                items.append({"itemnumber": item_number, "quantity": quantity})
            else:
                print("Invalid item number. Please select a valid item.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    
    order = {"Orderid": order_id, "items": items}
    orders.append(order)
    print(f"Order {order_id} created successfully!")

def generate_order_id():
    return "".join([str(random.randint(0, 9)) for _ in range(5)])

def view_orders():
    for order in orders:
        print(f"Order ID: {order['Orderid']}")
        total_cost = 0
        for item in order['items']:
            item_number = item['itemnumber']
            quantity = item['quantity']
            item_name, item_price = menu[item_number - 1]
            item_cost = item_price * quantity
            total_cost += item_cost
            print(f"{item_name} x{quantity} - ${item_cost}")
        print(f"Total Cost: ${total_cost}")
        print()

if __name__ == "__main__":
    while True:
        print("Options:")
        print("1. Create new order")
        print("2. View Orders")
        print("3. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            create_new_order()
        elif choice == '2':
            view_orders()
        elif choice == '3':
            break
        else:
            print("Invalid choice. Please select a valid option.")

    # Additional task (Optional): Save the order details to a file
    with open("orders.txt", "w") as file:
        for order in orders:
            file.write(str(order) + "\n")
```
